```python
import pandas as pd
import pylab as pl
import seaborn as sns

%pylab inline
```

```python
df = pd.read_csv('giraph-test.csv', index_col='Id')
df.fillna('S', inplace=True)
df
```

```python
df_success = df[(df["Status"] == "S")]
df_success
```

```python
def means(df_group, col):
    means = df_group[["Workers", col]].groupby("Workers").mean()
    return means.index.tolist(), means.values * 1.0 / 3600 / 1000

pl.figure(figsize=(10, 15))
for i, (mem, df_group) in enumerate(pd.groupby(df_success, "Mem")):
    
    pl.subplot(4, 1, i + 1)
    
    col = "Startup"
    color = "blue"
    pl.plot(df_group["Workers"], df_group[col] * 1.0 / 3600 / 1000, "o", color=color, label="")
    w, m = means(df_group, col)    
    pl.plot(w, m, "-", label=col, color=color)
    
    col = "Train"
    color = "green"
    pl.plot(df_group["Workers"], df_group[col] * 1.0 / 3600 / 1000, "o", color=color, label="")
    w, m = means(df_group, col)    
    pl.plot(w, m, "-", label=col, color=color)
    
    col = "Predict"
    color = "red"
    pl.plot(df_group["Workers"], df_group[col] * 1.0 / 3600 / 1000, "o", color=color, label="")
    w, m = means(df_group, col)    
    pl.plot(w, m, "-", label=col, color=color)
    
    col = "Total"
    color = "magenta"
    pl.plot(df_group["Workers"], df_group[col] * 1.0 / 3600 / 1000, "o", color=color, label="")
    w, m = means(df_group, col)    
    pl.plot(w, m, "-", label=col, color=color)
    
    pl.xlim(100, 600)
    pl.title("Memory: %d" % mem)
    pl.legend()
    
pl.show()
```
