```python
import bokeh.plotting as bp
import bokeh.models as bm
import datetime as dt
import numpy as np
import pandas as pd
import urlparse as up

import json
from collections import *

# bp.output_notebook()
```

```python
z = OrderedDict()
with open('/Users/anokhin/Desktop/sessions_chunk.txt') as f:
    for line in f:
        uid, js = line.split("\t")
        
        if uid not in z:
            z[uid] = []
        
        lines = z[uid]
        lines.append(json.loads(js))
        
        if len(z) > 20000:
            break

ind = 36
plot_id = z.items()[ind][0]
df = pd.DataFrame(z.items()[ind][1])
df
```

```python
# df["x"] = df["timestamp"].map(lambda ts: dt.datetime.fromtimestamp(ts - ts % 60))
# df["y"] = df["timestamp"].map(lambda ts: ts % 60)
start_minute_ts = df["timestamp"].min() - df["timestamp"].min() % 60
end_minute_ts = df["timestamp"].max() - df["timestamp"].max() % 60

df["y"] = (df["timestamp"].map(lambda ts: ts - ts % 60) - start_minute_ts) / 60
df["x"] = df["timestamp"].map(lambda ts: ts % 60)
df["url"] = df["hit"]
df["abbr"] = df["hit"].map(lambda url: up.urlparse(url).netloc.strip("www.")[0].upper())
df["email_color"] = df["email"].map(lambda i: "#339933" if pd.notnull(i) else "#888888")
df["okid_color"] = df["okid"].map(lambda i: "#FF6600" if pd.notnull(i) else "#888888")
df["vkid_color"] = df["vkid"].map(lambda i: "#0066FF" if pd.notnull(i) else "#888888")
df["text_color"] = df["gender"].map(lambda i: "white" if pd.isnull(i) else ("blue" if i else "red"))
df["color"] = "darkgray"
data_frame = df

links = []
referrers = {}
for i, row in df.sort(columns="timestamp", ascending=False, inplace=False).iterrows():
    if pd.notnull(row["hit"]) and row["hit"] in referrers:  
        for x_end, y_end in referrers[row["hit"]]:
            links.append(([row["x"], x_end], [row["y"], y_end]))
        del referrers[row["hit"]]
    if pd.notnull(row["referrer"]):
        endpoints = referrers.get(row["referrer"], [])
        endpoints.append((row["x"], row["y"]))
        referrers[row["referrer"]] = endpoints
```

```python
bp.output_server(plot_id)

def plot_arrow(link, pic):
    x, y = link
    pic.line(x, y, color='black')

p = bp.figure(
#     title="Timeline 1",
#     y_axis_type="datetime",
    tools="pan,reset,tap"
)

p.x_range = bm.Range1d(-1, 61)
p.y_range = bm.Range1d(data_frame["y"].max() + 1, data_frame["y"].min() - 1)

p.border_fill = "whitesmoke"

p.plot_height = int((data_frame["y"].max() - data_frame["y"].min()) * 60)
p.plot_width = 1200

p.xaxis[0].ticker = bm.FixedTicker(ticks=np.linspace(0, 60, 13))
p.xaxis.axis_line_width = 0
p.xaxis.axis_line_color = None

# p.xgrid.grid_line_color = None

p.xaxis.axis_label = "session seconds"
p.yaxis.axis_label = "session minutes"
# p.yaxis[0].ticker = bm.DatetimeTicker(desired_num_ticks=(data_frame["y"].max() - data_frame["y"].min()).seconds/60)

# p.xgrid.band_fill_alpha = 0.05
# p.xgrid.band_fill_color = "grey"

radius = 0.5
multi = 0.25

for link in links:
    plot_arrow(link, p)

s = bm.ColumnDataSource(data_frame)

email_annulus = bm.Annulus(
    x='x', y='y', 
    fill_color='email_color', 
    fill_alpha=1,
    line_color=None, 
    inner_radius=radius * (1.0 + 0 * multi),
    outer_radius=radius * (1.0 + 1 * multi)   
)
p.add_glyph(s, email_annulus, selection_glyph=email_annulus, nonselection_glyph=email_annulus)

okid_annulus = bm.Annulus(
    x='x', y='y', 
    fill_color='okid_color', 
    fill_alpha=1,
    line_color=None,     
    inner_radius=radius * (1.0 + 1 * multi),
    outer_radius=radius * (1.0 + 2 * multi)
)
p.add_glyph(s, okid_annulus, selection_glyph=okid_annulus, nonselection_glyph=okid_annulus)

vkid_annulus = bm.Annulus(
    x='x', y='y', 
    fill_color='vkid_color',
    fill_alpha=1,
    line_color=None, 
    inner_radius=radius * (1.0 + 2 * multi),
    outer_radius=radius * (1.0 + 3 * multi)
)
p.add_glyph(s, vkid_annulus, selection_glyph=vkid_annulus, nonselection_glyph=vkid_annulus)

visible_circle = bm.Circle(x='x', y='y', fill_color='color', fill_alpha=1.0, line_color=None, radius=radius)
r = p.add_glyph(s, visible_circle, selection_glyph=visible_circle, nonselection_glyph=visible_circle)

text = bm.Text(x='x', y='y', text='abbr', text_align='center', text_baseline='middle', text_color='text_color', text_font_size='8pt')
p.add_glyph(s, text, selection_glyph=text, nonselection_glyph=text)

taptool = p.select(type=bm.TapTool)
taptool.callback = bm.OpenURL(url="@url")
taptool.renderers = [r]

code = "source.set('selected', cb_data['index']);"
callback = bm.CustomJS(args={'source': s}, code=code)
p.add_tools(
    bm.HoverTool(
        tooltips=[("url", "@url"), ("VID", "@vid"), ("email", "@email"), ("OK", "@okid"), ("VK", "@vkid")], 
        callback=callback, 
        renderers=[r],
        mode='mouse', 
        line_policy='nearest',
        point_policy='snap_to_data'
    )
)

ann = bm.ColumnDataSource(data={
        'y': [0, (end_minute_ts - start_minute_ts) / 60], 
        'x': [0, 0], 
        'text': ['Session start %s' % dt.datetime.fromtimestamp(start_minute_ts), 'Session end %s' % dt.datetime.fromtimestamp(end_minute_ts)]
})
sb = bm.Text(x='x', y='y', text='text', text_align='left', text_baseline='bottom', text_color='black', text_font_size='10pt')
p.add_glyph(ann, sb, selection_glyph=sb, nonselection_glyph=sb)

bp.show(p)
```
