```python
import numpy as np
import pylab as pl
import seaborn as sns

%pylab inline
```

```python
N = 1000
Y = np.random.binomial(1, 0.5, N)

ind_pos = np.arange(N)[Y == 1]
ind_neg = np.arange(N)[Y == 0]

n_pos = ind_pos.shape[0]
n_neg = ind_neg.shape[0]

X = np.zeros((N, 2))
X[ind_pos, 0] = np.random.poisson(lam=1, size=n_pos)
X[ind_neg, 0] = np.random.poisson(lam=0.5, size=n_neg)

X[ind_pos, 1] = np.random.normal(1, 1, size=n_pos)
X[ind_neg, 1] = np.random.normal(5, 2, size=n_neg)

# print X, Y
```

```python
pl.figure(figsize=(20, 5))

pl.subplot(1, 3, 1)
pl.scatter(X[ind_pos, 0], X[ind_pos, 1], color='r', label='positive')
pl.scatter(X[ind_neg, 0], X[ind_neg, 1], color='b', label='negative')
pl.xlabel("$x_1$")
pl.ylabel("$x_2$")
pl.title("Data")
pl.legend()

pl.subplot(1, 3, 2)
pl.hist(X[ind_pos, 0], color='r', alpha=0.5, label='positive', bins=np.linspace(X[:, 0].min(), X[:, 0].max(), 20))
pl.hist(X[ind_neg, 0], color='b', alpha=0.5, label='negative', bins=np.linspace(X[:, 0].min(), X[:, 0].max(), 20))
pl.xlabel("$x_1$")
pl.ylabel("count")
pl.title("Histogram of $x_1$")
pl.legend()

pl.subplot(1, 3, 3)
pl.hist(X[ind_pos, 1], color='r', alpha=0.5, label='positive', bins=np.linspace(X[:, 1].min(), X[:, 1].max(), 20))
pl.hist(X[ind_neg, 1], color='b', alpha=0.5, label='negative', bins=np.linspace(X[:, 1].min(), X[:, 1].max(), 20))
pl.xlabel("$x_2$")
pl.ylabel("count")
pl.title("Histogram of $x_2$")
pl.legend()

pl.show()
```
